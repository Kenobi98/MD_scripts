```python
#Modifying the top file for domain swapping simulation
import numpy as np

with open ("GB1.27352.pdb.top","r") as top:
    top = top.readlines()

heads = ["[ atoms ]", "[ pairs ]", "[ bonds ]", "[ exclusions ]", "[ angles ]", "[ dihedrals ]", "[ system ]"]

#Adding 2nd chain atoms
n = len(top)   #length of the top file

#Calculating the no. of rows in a module
row = [];
for m in range(len(heads)-1):
    for i in range(len(top)):
        if heads[m] in top[i]:
            start = i+2
            rows = 0
            for j in range(start,len(top)):
                if heads[m+1] in top[j]:
                    break
                else:
                    rows += 1  
            row.append(rows)
        else:
            continue
```

```python
#Modifying the atoms section
w = open("symmetrized_gb1.top","w")
y = []
nxt = 0

for i in range(n):
    nxt += 1
    if heads[0] in top[i]:
        end = i+row[0]+1                 #Length of the atoms section
        w.writelines(top[i])             #Writing the section headers
        w.writelines(top[i+1])
        for j in range(i+2,end):
            nxt += 1
            a = top[j].split()
            a[0] = str(int(a[0])+436)    #Adding 2nd chain atoms
            a[2] = str(int(a[2])+56)
            a[5] = str(int(a[5])+436)

            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(2 + len(a[1]))
            a[2] = a[2].rjust(12 - len(a[1]))
            a[3] = a[3].rjust(5)
            a[4] = a[4].rjust(2 + len(a[4]))
            a[5] = a[5].rjust(5 + (7 - len(a[4])))
            a[6] = a[6].rjust(3 + len(a[6]))
            a[7] = a[7].rjust(3 + len(a[7]))

            b = a[0] + a[1] + a[2] + a[3] + a[4] + a[5] + a[6] + a[7] + "\n"            
            y.append(str(b))
            w.writelines(top[j])
        break
    else:
        w.writelines(top[i])

w.writelines(y[:])


#Modifying the pairs section

y = []; z1 = []; z2 = []
nxt += 1

for i in range(nxt,n):
    nxt += 1
    if heads[1] in top[i]:
        end = i+row[1]+1
        w.writelines(top[i])
        w.writelines(top[i+1])
        for j in range(i+2,end):
            nxt += 1
            a = top[j].split()

            a[0] = str(int(a[0])+436)   #Adding 2nd chain 
            a[1] = str(int(a[1])+436)
            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(4 + len(a[1]))

            b = "-" + a[0]+ a[1] + "  " + a[2] + "   " + a[3] + "   " + a[4] +  "\n"
            y.append(str(b))

            a = top[j].split()
            a[1] = str(int(a[1])+436)  #Adding interchain contact I-J'
            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(7)
            b = ":" + a[0] + a[1] + "  " + a[2] + "   " + a[3] + "   " + a[4] +  "\n"
            z1.append(str(b))

            a = top[j].split()
            a[0] = str(int(a[0])+436)  #Adding interchain contact I'-J
            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(7)
            b = "/" + a[0] + a[1] + "  " + a[2] + "   " + a[3] + "   " + a[4] +  "\n"
            z2.append(str(b))

            w.writelines(top[j])
        break
    else:
        w.writelines(top[i])

w.writelines(y[:])
w.writelines(z1[:])
w.writelines(z2[:])


#Modifying the bonds section
y = []; z1 = []; z2 = [];
nxt += 1

for i in range(nxt,n):
    nxt += 1
    if heads[2] in top[i]:
        end = i+row[2]+1
        w.writelines(top[i])
        w.writelines(top[i+1])
        for j in range(i+2,end):
            nxt += 1
            a = top[j].split()

            a[0] = str(int(a[0])+436)  
            a[1] = str(int(a[1])+436)
            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(6)

            b = "-" + a[0] + a[1] + "  " + a[2] + "   " + a[3] + "   " + a[4] +  "\n"

            y.append(str(b))
            w.writelines(top[j])
        break
    else:
        w.writelines(top[i])

w.writelines(y[0:])


#Modifying the exclusions section
y = []; z1 = []; z2 = [];
nxt += 1

for i in range(nxt,n):
    nxt += 1
    if heads[3] in top[i]:
        end = i+row[3]+1
        w.writelines(top[i])
        w.writelines(top[i+1])
        for j in range(i+2,end):
            nxt += 1
            a = top[j].split()
            
            a[0] = str(int(a[0])+436)
            a[1] = str(int(a[1])+436)    #Adding 2nd chain exclusions
            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(7)
            b = "-" + a[0] + a[1] + "\n"
            y.append(b)

            a = top[j].split()
            a[1] = str(int(a[1])+436)    #Adding I-J'exclusions
            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(7)
            b = ":" + a[0] + a[1] + "\n"
            z1.append(b)

            a = top[j].split()
            a[0] = str(int(a[0])+436)    #Adding I'-J exclusions
            a[0] = a[0].rjust(6)
            a[1] = a[1].rjust(7)
            b = "/" + a[0] + a[1] + "\n"
            z2.append(b)

            w.writelines(top[j])
        break
    else:
        w.writelines(top[i])

w.writelines(y[0:])
w.writelines(z1[0:])
w.writelines(z2[0:])


#Modifying the angles section
y = []
nxt += 1

for i in range(nxt,n):
    nxt += 1
    if heads[4] in top[i]:
        end = i+row[4]+1
        w.writelines(top[i])
        w.writelines(top[i+1])
        for j in range(i+2,end):
            nxt += 1
            a = top[j].split()

            a[0] = str(int(a[0])+436)  
            a[1] = str(int(a[1])+436)
            a[2] = str(int(a[2])+436)
            a[0] = a[0].rjust(3 + len(a[0]))
            a[1] = a[1].rjust(4 + len(a[1]))
            a[2] = a[2].rjust(4 + len(a[2]))

            b = "-" + a[0] + a[1] + a[2] + "  " + a[3] + "   " + a[4] + "  " + a[5] + "\n"

            y.append(str(b))
            w.writelines(top[j])
        break
    else:
        w.writelines(top[i])

w.writelines(y[0:])


#Modifying the dihedral section
y = []
nxt += 1

for i in range(nxt,n):
    nxt += 1
    if heads[5] in top[i]:
        end = i+row[5]+1
        w.writelines(top[i])
        w.writelines(top[i+1])
        for j in range(i+2,end):
            nxt += 1
            a = top[j].split()

            a[0] = str(int(a[0]) + 436)  
            a[1] = str(int(a[1]) + 436)
            a[2] = str(int(a[2]) + 436)
            a[3] = str(int(a[3]) + 436)
            a[0] = a[0].rjust(3 + len(a[0]))
            a[1] = a[1].rjust(4 + len(a[1]))
            a[2] = a[2].rjust(4 + len(a[2]))
            a[3] = a[3].rjust(4 + len(a[3]))

            if len(a) == 7:
                a.append(" ")


            b = "-" + a[0] + a[1] + a[2] + a[3] + "  " + a[4] + "  " + a[5] +  "  " + a[6] + " " + a[7] + "\n"

            y.append(str(b))
            w.writelines(top[j])
        break
    else:
        w.writelines(top[i])

w.writelines(y[0:])


#A glorious ending
for i in range(nxt,n):
    nxt += 1
    if heads[6] in top[i]:
        w.writelines("\n")
        w.writelines(top[i:n])

w.close()
```
