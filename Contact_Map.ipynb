```python
import numpy
from matplotlib import pyplot as plt

contacts = open ("Contacts.txt", "r")
cont = contacts.readlines()
```

```python
"""Plotting Contact Map"""

x = []
y = []

split = []

for i in range(len(cont)):
    split = cont[i].split()
    x.append(int(split[0]))
    y.append(int(split[1]))
    x.append(int(split[1]))
    y.append(int(split[0]))
fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(x,y,  c = "blue", linewidth = 0.5, s = 10)
plt.xlim(0, 440)
plt.ylim(0, 440)
plt.title("Contact Map")
plt.xlabel("Atom no. (i)")
plt.ylabel("Atom no. (j)")
plt.grid(True)
ax.set_aspect('equal', adjustable='box')

#plt.show()
plt.savefig("Contact_Map.png", dpi = 300, format = "png")
```
