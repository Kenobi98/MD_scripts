```python
import numpy as np
import pandas as pd
import math

file = open ("GB1.27352.pdb.top","r")
top = file.readlines()      #Reading the top file

gb_pdb = open ("mod.pdb", "r")
pdb = gb_pdb.readlines()    #Reading the PDB file

trj = open ("traj_short.pdb","r")
traj = trj.readlines()      #Reading the trajectory file
```

```python
"""Counting the no. of rows in a module"""
rows = 0

for i in range(len(top)):
    if "[ pairs ]" in top[i]:
        start = i+2
        for j in range(start,len(top)):
            if "[ bonds ]" in top[j]:
                break
            else:
                rows += 1         
    else:
        continue

"""Selecting i-j pairs from top file"""
        
atm1 = []
atm2 = []
end = start+rows-1
for i in range(start,end):
    line = top[i].split()
    atm1.append(line[0])
    atm2.append(line[1])          
print(atm1[0])
```

```python
"""Function to convert pdb to DataFrame"""

def pdb_df2 (p):
    n = 436
    y = []
    atm_no = []
    x_cord = []
    y_cord = []
    z_cord = []

    for i in range(n):
        y = p[i].split()
        atm_no.append(y[1])
        x_cord.append(y[5])
        y_cord.append(y[6])
        z_cord.append(y[7])
        
    atm_no = pd.DataFrame(data = atm_no)
    x_cord = pd.DataFrame(data = x_cord)
    y_cord = pd.DataFrame(data = y_cord)
    z_cord = pd.DataFrame(data = z_cord)
    
    return (atm_no,x_cord,y_cord,z_cord)

atm_no, x_cord, y_cord, z_cord = pdb_df2(pdb)
data = pd.DataFrame()
data["atm_no"] = atm_no[0].values
data["X"] = x_cord[0].values
data["Y"] = y_cord[0].values
data["Z"] = z_cord[0].values
```

```python
"""Function to calculate distance between i-j pairs in native state"""
native_dist = pd.DataFrame()


a = []
b = []
c = []
x1 = 0; x2 = 0; y1 = 0; y2 = 0; z1 = 0; z2 = 0;

for i in range(len(atm1)):
    for j in range(len(data)):
        if data.iloc[j,0] == atm1[i]:
            x1 = data.iloc[j,1]
            y1 = data.iloc[j,2]
            z1 = data.iloc[j,3]
            
            
        if data.iloc[j,0] == atm2[i]:
            x2 = data.iloc[j,1]
            y2 = data.iloc[j,2]
            z2 = data.iloc[j,3]
            
    x = float(x2) - float(x1)
    y = float(y2) - float(y1)
    z = float(z2) - float(z1)
    dist = math.sqrt(x**2 + y**2 + z**2)
    a.append(atm1[i])
    b.append(atm2[i])
    c.append(dist)

a = pd.DataFrame(data = a)
b = pd.DataFrame(data = b)
c = pd.DataFrame(data = c)
 
native_dist["i"] = a[0].values
native_dist["j"] = b[0].values
native_dist["distance"] = c[0].values
```

```python
"""Preparing the Q and Qi files data"""
ai = []
aj = []
q_count = []
ii = 0
call = []

for i in range(len(traj)):     #
    if "MODEL" in traj[i]:
        atm_n = []
        xcord = []
        ycord = []
        zcord = []

        for j in range(i+1,len(traj)):
            if "TER" in traj[j]:
                break
            else:
                y = traj[j].split()
                atm_n.append(y[1])
                xcord.append(y[5])
                ycord.append(y[6])
                zcord.append(y[7])

        atm_n = pd.DataFrame(data = atm_n)
        xcord = pd.DataFrame(data = xcord)
        ycord = pd.DataFrame(data = ycord)
        zcord = pd.DataFrame(data = zcord)
            
        data_trj = pd.DataFrame()
        data_trj["atm_no"] = atm_n[0].values
        data_trj["X"] = xcord[0].values
        data_trj["Y"] = ycord[0].values
        data_trj["Z"] = zcord[0].values

        a = []
        b = []
        c = []
        x1 = 0; x2 = 0; y1 = 0; y2 = 0; z1 = 0; z2 = 0;

        for m in range(len(atm1)):
            for n in range(len(data_trj)):
                if data_trj.iloc[n,0] == atm1[m]:
                    x1 = data_trj.iloc[n,1]
                    y1 = data_trj.iloc[n,2]
                    z1 = data_trj.iloc[n,3]

                if data.iloc[n,0] == atm2[m]:
                    x2 = data_trj.iloc[n,1]
                    y2 = data_trj.iloc[n,2]
                    z2 = data_trj.iloc[n,3]

            x = float(x2) - float(x1)
            y = float(y2) - float(y1)
            z = float(z2) - float(z1)
            dist = math.sqrt(x**2 + y**2 + z**2)
            a.append(atm1[m])
            b.append(atm2[m])
            c.append(dist)
        a = pd.DataFrame(data = a)
        b = pd.DataFrame(data = b)
        c = pd.DataFrame(data = c)
        
        traj_dist = pd.DataFrame()
        traj_dist["i"] = a
        traj_dist["j"] = b
        traj_dist["distance"] = c
        
        count = 0
        
        for o in range(len(traj_dist)):        #Comparing interaction in native state and trajectory
            if traj_dist.iloc[o,2]<=(native_dist.iloc[o,2]*1.5):
                count += 1
                call.append(count)
                ai.append(traj_dist.iloc[o,0])
                aj.append(traj_dist.iloc[o,1])

        q_count.append(count)
        ii += 1
        print("Run ", i, " --> ",ii)
                
    else:
        continue
```

```python
"""Writing in the Q file"""
string = []
for i in range(len(q_count)):
    a = str(q_count[i])+"\n"
    string.append(a)

Q = open("Q.txt","w")
Q.writelines(string)
Q.close()
```
